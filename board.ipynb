$$
f(x) = x^5 + x - 3
$$

```python
import matplotlib.pyplot as plt
import numpy
from numpy import arange

r = arange(-2, 2, 0.0001)

plt.plot(r, list(map(lambda x: x**5 + x - 3, r)))
plt.show()
```
